```python
import xarray as xr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from xmca.xarray import xMCA
from pyEOF import *
```

```python
min_temp = xr.open_dataset('tr_min_ann_full.nc')['tasmin']
```

```python
min_temp
```

```python
b = min_temp.isel(time=6)
b.plot()
```

```python
pca = xMCA(min_temp)                       # PCA on field A
pca.solve()        # True for complex PCA
# pca.rotate(10)                     # optional; Varimax rotated solution
                                    # using 10 first EOFs
eigenvalues = pca.singular_values() # singular vales = eigenvalues for PCA
pcs         = pca.pcs()             # Principal component scores (PCs)
eofs        = pca.eofs()            # spatial patterns (EOFs)

pca.plot(mode=1)                    # plot mode 1
```

```python
pca.plot(mode=2)
```

```python
pca.plot(mode=3)
```
